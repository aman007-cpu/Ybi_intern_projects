```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'https://github.com/ybifoundation/Dataset/raw/main/Bike%20Prices.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.describe()
```

```python
df[['Brand']].value_counts()
```

```python
df[['Model']].value_counts()
```

```python
#Get Column Names
```

```python
df.columns
```

```python
#Get Shape of DataFrame
```

```python
df.shape
```

```python
# Get Encoding of Categorical Features
```

```python
df.replace({'Seller_Type':{'Individual':0,'Dealer':1}},inplace=True)
```

```python
df.replace({'Owner':{'1st owner':0,'2nd owner':1, '3rd owner':2, '4th owner':3}},inplace=True)
```

```python
# Definey(dependent or label or target variable)andX(independent or features or attribute Variable)
```

```python
y = df['Selling_Price']
```

```python
y.shape
```

```python
y
```

```python
X=df[['Year','Seller_Type','Owner','KM_Driven','Ex_Showroom_Price']]
```

```python
X.shape
```

```python
X
```

```python
# Get train test split
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=12529)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
# get model train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
# get model prediction
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
# get model evaluation
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
# Get Visualization of Actual Vs Predicted Results
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Preicted Price")
plt.show()
```
